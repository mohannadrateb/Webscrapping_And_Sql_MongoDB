```python
from bs4 import BeautifulSoup
import requests
```

```python
url = 'https://www.datajob.io/'
page = requests.get(url)
soup = BeautifulSoup(page.text,'html')

```

```python
divs = soup.find_all('div', class_='content_card-block')
```

```python

job_names =[]
company_names=[]
job_descs =[]
job_infos= []
for div in divs:
    job_info_array =[]
    nested_div_job_name = div.find('div', class_='content_card-title')
    nested_div_job_name = nested_div_job_name.text.strip() 
    nested_div_company_name = div.find('div', class_='content_card-company-name')
    nested_div_company_name = nested_div_company_name.text.strip()
    nested_div_job_desc = div.find('div', class_='content-card-div-job-desc').text.strip()
    nested_div_job_info = div.find_all('div', class_='content_card-location-ff-salary')
    for job in nested_div_job_info:
        nested_div_job_info = job.find_all('div', class_='content_card-info-text')
        for job_info in nested_div_job_info:
            #print(job_info.text.strip())
            job_info_array.append(job_info.text.strip())
    
    #print(nested_div_job_name)
    #print(nested_div_company_name)
    #print(nested_div_job_desc)
    #print(nested_div_job_info)
    #print(nested_div_job_req)
    job_names.append(nested_div_job_name)
    company_names.append(nested_div_company_name )
    job_descs.append(nested_div_job_desc)
    job_infos.append(job_info_array)




    #break
```

```python
print(job_names)
print("--------------------")
print(company_names)
print('-----------------------')
print(job_descs)
print('-------------------')
print(job_infos)
```

```python
import pandas as pd
df = pd.DataFrame(columns = ['company_name','job_name','job_desc','job_req_and_info'])
```

```python
df['company_name'] = company_names
df['job_name'] = job_names
df ['job_desc'] = job_descs
df['job_req_and_info'] =job_infos
```
